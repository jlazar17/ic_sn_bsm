## Put these into a language that `SNEWPY` understands

`SNEWPY` wants the fluxes parametrized using the following 13 parameters:

<!-- 1. $t$
2. $L_{\nu_{e}}$
3. $L_{\bar{\nu}_{e}}$
4. $L_{\nu_{x}}$
5. $E_{\nu_{e}}$
6. $E_{\bar{\nu}_{e}}$
7. $E_{\nu_{x}}$
8. $E^{{\rm{RMS}}}_{\nu_{e}}$
9. $E^{{\rm{RMS}}}_{\bar{\nu}_{e}}$
10. $E^{{\rm{RMS}}}_{\nu_{x}}$
11. $\alpha_{\nu_{e}}$
12. $\alpha_{\bar{\nu}_{e}}$
13. $\alpha_{\nu_{x}}$ -->
1. `TIME`
2. `L_NU_E`
3. `L_NU_E_BAR`
4. `L_NU_X`
5. `E_NU_E`
6. `E_NU_E_BAR`
7. `E_NU_X`
8. `RMS_NU_E`
9. `RMS_NU_E_BAR`
10. `RMS_NU_X`
11. `ALPHA_NU_E`
12. `ALPHA_NU_E_BAR`
13. `ALPHA_NU_X`

Where `L_X` is the luminosity of the flux for neutrino species `X` expressed in ergs per seconds, `E_X` is the mean energy of neutrino species `X` expressed in MeV, `RMS_X` is the root-mean-square of distribution for species `X` expressed in MeV, and `ALPHA_X` is a shape parameter for neutrino species `X`.
Between the [`SNEWPY` paper](https://arxiv.org/abs/2109.08188) and another [older paper](https://arxiv.org/abs/astro-ph/0208035), I think the following formulae for these hold

$$
L_{x} = 10^{51} \times \int {\rm{d}}E \, E \, \Phi_{x}
$$

$$
\left<E_{x}\right> = \frac{\int {\rm{d}}E \, E \, \Phi_{x}}{\int {\rm{d}}E \, \Phi_{x}}
$$

$$
E^{\rm{RMS}}_{x} = \sqrt{\frac{\int {\rm{d}}E \, E^3 \, \Phi_{x}}{\int {\rm{d}}E \, E \, \Phi_{x}}} = \sqrt{\frac{\left<E^{3}\right>}{\left<E\right>}}
$$

$$
\alpha_{x} = \frac{2\left<E\right>^{2} - \left<E^2\right>}{\left<E^2\right>- \left<E\right>^2}
$$

Where $\Phi_{x} = \frac{{\rm{d}}N_{x}}{{\rm{d}}E {\rm{d}}t}$.

The [`SNEWPY` paper](https://arxiv.org/abs/2109.08188) then claims that this can be converted to a flux via the folowing prescription.

$$
\Phi(E_{\nu}) = \frac{L_{\nu}}{\left<E_{\nu}\right>} \frac{\left(\alpha+1\right)^{\alpha+1}}{\Gamma\left(\alpha+1\right)}\left(\frac{E_{\nu}}{\left<E_{\nu}\right>}\right)^{\alpha}\exp\left[\frac{-\left(\alpha+1\right)E_{\nu}}{\left<E_{\nu}\right>}\right]
$$

$$
\Phi(E_{\nu}) = \frac{L_{\nu}}{\left<E_{\nu}\right>} \frac{\left(\alpha+1\right)^{\alpha+1}}{\Gamma\left(\alpha+1\right)}\left(\frac{E_{\nu}}{\left<E_{\nu}\right>}\right)^{\alpha}\exp\left[-\left(\alpha+1\right)\frac{E_{\nu}}{\left<E_{\nu}\right>}\right]
$$

## Now we can throw this through `SNEWPY` and `ASTERIA`

Let's do a quick check to make sure that I know how to use this.
According to some people that I have been chatting with, a SN with a mass of 27 solar masses at a distance of 10 kpc should produce $~7\times10^5$ hits in the first ten seconds when considering the Sukhbold model.
Let's make sure that we can reproduce that.

```python
dt = 0.5 * u.s
sim_kwargs = {
    # "distance": 1 * u.m,
    "Emin": 0*u.MeV,
    "Emax": 400*u.MeV,
    "dE": 0.5*u.MeV,
    "tmin": -1*u.s,
    "tmax": 20*u.s,
    "dt": 1*u.ms,
    "mixing_scheme": 'AdiabaticMSW',
    "hierarchy": 'normal',
}

model = {
    'name' : 'Sukhbold_2015',
    'param' : ccsn.Sukhbold_2015.get_param_combinations()[0]

}

sim = Simulation(model=model, distance=10.0 * u.kpc, **sim_kwargs)
sim.run()
sim.rebin_result(dt)
t, hits = sim.detector_signal(dt)
m = t.value < 10
print(f"We saw {hits[m].sum()} in the first ten seconds.")
```

```python
tmins = np.linspace(0, 20, 201)
tmaxs = np.linspace(0, 20, 201)

significance = np.full(tmins.shape + tmaxs.shape, np.nan)

f = 6

for idx, tmin in enumerate(tmins):
    for jdx, tmax in enumerate(tmaxs):
        m = np.logical_and(tmin < bsm_t.value, bsm_t.value < tmax)
        all_cts = (bsm_hits[m]/f + sm_hits[m] + sm_bg[m]).sum()
        if all_cts==0:
            continue
        bsm_cts = bsm_hits[m].sum() / f        
        significance[idx, jdx] = bsm_cts / np.sqrt(all_cts)

        
fig, ax = plt.subplots(figsize=(6,5))
[x.set_linewidth(1.5) for x in ax.spines.values()]

im = ax.imshow(
    significance[::-1],
    extent=[0, 20, 0, 20],
    aspect="auto"
)

ax.set(
    xlim=(0, 20),
    ylim=(0, 20),
    xlabel=r"$t_{\rm{max}}~\left[{\rm{sec}}\right]$",
    ylabel=r"$t_{\rm{min}}~\left[{\rm{sec}}\right]$",
)

cbar = plt.colorbar(im, label=r"$\frac{N_{\rm{BSM}}}{\sqrt{N_{\rm{tot.}}}}$")

plt.show()
```
